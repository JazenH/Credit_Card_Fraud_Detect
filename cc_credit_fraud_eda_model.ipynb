# Before we build models, we should consider how to deal with the imbalance. 

The data has 284315 non-fraud and 492 fraud records, which is highly imbalanced. Building model based on the skewness is not recommended. We first want to the undersampling, constructing multiple samples each has balanced data.

```python
# First shuffle the data before creating the subsamples, so the time is not ordered

df = df.sample(frac=1)
```

```python
fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492] #We shuffled the dataframe before, here the sampling become simple
```

```python
even_df = pd.concat([fraud_df, non_fraud_df])

# Shuffle dataframe rows
new_df = even_df.sample(frac=1, random_state=42)

new_df.shape
```

```python
print(new_df['Class'].value_counts()/len(new_df))
```

The subsample is evenly distributed

```python
new_df_corr=new_df.corr()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(new_df_corr,linewidths=1,cmap='Reds')
plt.show()
```

Before under-sampling non-fraud data is overwhelmingly more the the fraud cases. After down-sampling we can see that there are some correlation between features the fraud transactions.

# Modeling

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
%matplotlib inline

scores =[]
for k in range(1, 200):
    rfc = RandomForestClassifier(n_estimators=k)
    rfc.fit(X_train, y_train)
    y_pred = rfc.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))




plt.plot(range(1, 200), scores)
plt.xlabel('Value of n_estimators for Random Forest Classifier')
plt.ylabel('Testing Accuracy')
```

Here, it is visible that a high value for n_estimators will give a good acuracy score, but it is fluctuating randomly in the curve even for nearby values of n_estimators, and higher n(>75) will not give noticablly better score.

## More details

### A. The confusion matrix

```python
from sklearn.metrics import confusion_matrix
rf_clf = RandomForestClassifier(n_estimators=100, oob_score=False)
rf_clf.fit(X_train, y_train)

y_train_pred = rf_clf.predict(X_train)
y_test_pred = rf_clf.predict(X_test)

CM_train = confusion_matrix(y_train, y_train_pred)
CM_test  = confusion_matrix(y_test, y_test_pred)
```

```python
print(f"The confusion_matrix of train set is:\n {CM_train}\n")
print(f"The confusion_matrix of test set is:\n {CM_test}\n")
```
